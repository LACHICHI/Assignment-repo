<font color="red"><h2> Method and Results </h2></font>

- Solve the question here and discuss the results <br />


```python


import numpy as np
import matplotlib.pyplot as plt

class Potential:
    def __init__(self, a, b, V):

        self.a = a
        self.b = b
        self.V = V

    def V_func(self, x):
        if 0 <= x <= self.a:
            return 0
        elif self.a < x <= self.b:
            return self.V
        else:
            return np.inf

class Schrodinger:

    def __init__(self, potential_well, N=1000):
        self.potential_well = potential_well
        self.N = N
        self.dx = potential_well.b / N
        self.x = np.linspace(0, potential_well.b, N)
        self.V_arr = np.array([potential_well.V_func(xi) for xi in self.x])

    def solve(self):


        H = np.zeros((self.N, self.N))
        for i in range(self.N):
            H[i, i] = 2.0 / self.dx**2 + self.V_arr[i]
            if i > 0:
                H[i, i-1] = -1.0 / self.dx**2
            if i < self.N-1:
                H[i, i+1] = -1.0 / self.dx**2
        E, psi = np.linalg.eigh(H)
        self.E = E
        self.psi = psi

    def PlotEigenvalues(self):
        plt.plot(range(1, self.N+1), self.E, 'o')
        plt.xlabel('Eigenvalue index')
        plt.title('Energy Eigenvalues')
        plt.ylabel('Energy')
        plt.show()

    def PlotWavefunctions(self, num=5):
        for i in range(num):
            plt.plot(self.x, self.psi[:, i], label=f'Eigenfunction {i+1}')
        plt.xlabel('Position ')
        plt.ylabel('Wavefunction')
        plt.legend()
        plt.show()



well1 = Potential(1.0, 2.0, 10.0)
solver1 = Schrodinger(well1)
solver1.solve()
solver1.PlotEigenvalues()
solver1.PlotWavefunctions()
```
